```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma = 'auto'),
        'params': {
            'C': [1,10,20,30],
            'kernel':['rbf','linear','sigmoid','poly'],
            'degree':[1,2,3]
        }
    },
    
    'random_forest':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators': [1,20,50,100],
            'criterion':['gini','entropy'],
            'max_features' : ["auto", "sqrt", "log2"]
        }
    },
    
    'logistic_regression':{
        'model': LogisticRegression(),
        'params':{
            'solver': ['liblinear','lbfgs'],
            'C':[1,5,10]
        }
    },
    
    'gausian_bayes':{
        'model':GaussianNB(),
        'params':{
            
        }
    },
    
    'multinomila_naive_bayes':{
        'model':MultinomialNB(),
        'params':{
            'alpha':[1,2]
        }
    },
    
    'decision_tree_classifier':{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion':['gini','entropy'],
            'splitter':['best','random']
        }
    }
}
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
score = []
score2 = []

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(digits.data,digits.target)
    
    score.append([
        model_name,
        clf.best_params_,
        clf.best_score_
    ])
    
    clfr = RandomizedSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False,n_iter=3)
    clfr.fit(digits.data,digits.target)
    score2.append([
        model_name,
        clfr.best_params_,
        clfr.best_score_
    ])
```

```python
import pandas as pd

df = pd.DataFrame(score,columns=['model','best_params','best_score'])
df
```
