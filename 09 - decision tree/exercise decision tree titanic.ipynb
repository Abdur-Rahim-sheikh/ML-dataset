```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.sample(1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for name in df:
    pd.crosstab(df[name][:100],df.Survived[:100]).plot(kind='bar')
```

```python
df2 = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Fare','Cabin'],axis='columns')
df2.head(5)
```

```python
# df2 = pd.get_dummies(data=df2,columns=['Sex','Embarked'],drop_first=True)
df2.Age.fillna(df2.Age.mean(),inplace = True)
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df2.drop(['Survived'],axis='columns'),df2.Survived)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```
